This notebook computes average metrics from the test-set evaluations for the models predicting forces, contact classification, and force distribution.

# Input model name here

```python
MODEL_NAME = 'with_physicsinputs_lstm_newoutputs_transformfix_rawdist_tunedparams_rotonly'
```

# Load libraries and set up

```python
# Mount Drive to access data.
from google.colab import drive
drive.mount('/content/drive')
```

```python
import numpy as np
import os
import pandas as pd
```

# Compute average metrics

```python
# Load regression metrics sheets.
cont_acl  = pd.read_excel(os.path.join('/content/drive/My Drive/CS230_FinalReport/reports/metrics/',MODEL_NAME, 'rmses_acl.xlsx'), index_col = 0)
cont_oagr = pd.read_excel(os.path.join('/content/drive/My Drive/CS230_FinalReport/reports/metrics/',MODEL_NAME, 'rmses_oagr.xlsx'), index_col = 0)
```

```python
# Get average R/L GRFs.
print(f"R/L GRF: {np.round(np.mean( [cont_acl.iloc[:, 0:8].mean().mean(), cont_oagr.iloc[:, 0:8].mean().mean()] ),2)}")

# Get average total GRFs.
print(f"tot GRF: {np.round(np.mean( [cont_acl.iloc[:, -4:].mean().mean(), cont_oagr.iloc[:, -4:].mean().mean()] ),2)}")
```

```python
# Load classification metrics sheets.
class_acc = pd.read_excel(os.path.join('/content/drive/My Drive/CS230_FinalReport/reports/metrics/',MODEL_NAME, 'contact_class_accs.xlsx'), index_col = 0)
```

```python
# Get average of R and L legs
print(f"class: {np.round(class_acc.iloc[0,:].mean(), 2)}")
```

```python
# Load distribution metrics sheets.
dist_acl  = pd.read_excel(os.path.join('/content/drive/My Drive/CS230_FinalReport/reports/metrics/',MODEL_NAME, 'rmses_acl_dist.xlsx'), index_col = 0)
dist_oagr = pd.read_excel(os.path.join('/content/drive/My Drive/CS230_FinalReport/reports/metrics/',MODEL_NAME, 'rmses_oagr_dist.xlsx'), index_col = 0)
```

```python
# Get total average of both R and L distributions.
print(f"dist: {np.round( np.mean([dist_acl.mean().mean(), dist_oagr.mean().mean()]) ,2)}")
```
