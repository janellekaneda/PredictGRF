# Input parameters here

```python
# # PARAMETERS # #
BASE_DIR = '/content/drive/My Drive/CS230_FinalReport/' # your base directory
EXP_TYPE = 'with_physicsinputs'

MODEL_FILENAME = '/content/drive/My Drive/CS230_FinalReport/models/' + EXP_TYPE + '_lstm_newoutputs_transformfix_rawdist_tunedparams_rotonly' # where you want to save the model to
SEED = 43
# # # # # # # # # #
```

# Load libraries and other set up

```python
from tensorflow import keras
import numpy as np
import os
```

```python
# Mount Drive to access data.
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Set the random seed
keras.utils.set_random_seed(SEED)
```

```python
# Load in data to predict on (dev or test sets).
datadir = os.path.join(BASE_DIR, 'data', 'processed', 'normalized')

X_test = np.load(file=os.path.join(datadir, EXP_TYPE, 'X_test_norm_100Hz_RawDist.npy'), allow_pickle=True)
```

# Gather predictions for later visualization and analysis

```python
# Load saved model.
saved_model = keras.models.load_model(MODEL_FILENAME, compile = False)

# Gather all predictions on test set using the saved model.
preds = saved_model.predict(X_test)
```

```python
# Save the preds.
outnames = ['testset_predictions_cont.npy', 'testset_predictions_class.npy', 'testset_predictions_dist.npy']
predsdir = os.path.join(MODEL_FILENAME, 'testset_predictions.npy')
for i, out in enumerate(preds):
  np.save(os.path.join(MODEL_FILENAME, outnames[i]), out)
```
